1. Create a `matrix` class with the following properties:
    * It can be initialized in 2 ways:
        1. with arguments `n` and `m`, the size of the matrix. A newly instanciated matrix will contain all zeros.
        2. with a list of lists of values. Note that since we are using lists of lists to implement matrices, it is possible that not all rows have the same number of columns. Test explicitly that the matrix is properly specified.
    * Matrix instances `M` can be indexed with `M[i][j]` and `M[i,j]`.
    * Matrix assignment works in 2 ways:
        1. If `M_1` and `M_2` are `matrix` instances `M_1=M_2` sets the values of `M_1` to those of `M_2`, if they are the same size. Error otherwise.
        2. In example above `M_2` can be a list of lists of correct size.


```python
class Matrix:
    def __init__(self, n, m=None):
        if isinstance(n, list):  # Initialize with a list of lists
            self.data = n
            self.rows = len(n)
            self.cols = len(n[0]) if n else 0
            if any(len(row) != self.cols for row in n):
                raise ValueError("All rows must have the same number of columns")
        else:  # Initialize with dimensions
            self.rows = n
            self.cols = m
            self.data = [[0] * m for _ in range(n)]

    def __getitem__(self, index):
        if isinstance(index, tuple):
            i, j = index
            return self.data[i][j]
        return self.data[index]

    def __setitem__(self, index, value):
        if isinstance(index, tuple):
            i, j = index
            self.data[i][j] = value
        else:
            self.data[index] = value

    def __repr__(self):
        return '\n'.join(str(row) for row in self.data)
    
    def assign(self, other):
        if isinstance(other, Matrix):
            if self.rows == other.rows and self.cols == other.cols:
                self.data = [row[:] for row in other.data]
            else:
                raise ValueError("Matrix sizes do not match")
        elif isinstance(other, list):
            if len(other) == self.rows and all(len(row) == self.cols for row in other):
                self.data = [row[:] for row in other]
            else:
                raise ValueError("List dimensions do not match matrix size")
        else:
            raise TypeError("Assignment must be from a Matrix or a list of lists")

# Testing the implementation
if __name__ == "__main__":
    print("Initializing a 3x3 zero matrix:")
    M1 = Matrix(3, 3)
    print(M1, "\n")
    
    print("Initializing a matrix with a list of lists:")
    M2 = Matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
    print(M2, "\n")
    
    print("Testing indexing:")
    print("M2[1,2] =", M2[1, 2])
    print("M2[2] =", M2[2], "\n")
    
    print("Testing assignment:")
    M1.assign(M2)
    print("M1 after assignment:")
    print(M1, "\n")
    
    print("Testing element modification:")
    M1[0, 0] = 10
    print(M1, "\n")
    
    print("Testing invalid assignment:")
    try:
        M1.assign([[1, 2], [3, 4]])
    except ValueError as e:
        print("Error:", e)
```

    * `row(n)` and `column(n)`: that return the nth row or column of the matrix M as a new appropriately shaped matrix object.

```python
class Matrix:
    def __init__(self, n, m=None):
        if isinstance(n, list):  # Initialize with a list of lists
            self.data = n
            self.rows = len(n)
            self.cols = len(n[0]) if n else 0
            if any(len(row) != self.cols for row in n):
                raise ValueError("All rows must have the same number of columns")
        else:  # Initialize with dimensions
            self.rows = n
            self.cols = m
            self.data = [[0] * m for _ in range(n)]

    def __getitem__(self, index):
        if isinstance(index, tuple):
            i, j = index
            if isinstance(i, slice) or isinstance(j, slice):
                return Matrix([row[j] for row in self.data[i]])
            return self.data[i][j]
        return self.data[index]

    def __setitem__(self, index, value):
        if isinstance(index, tuple):
            i, j = index
            self.data[i][j] = value
        else:
            self.data[index] = value

    def __repr__(self):
        return '\n'.join(str(row) for row in self.data)
    
    def assign(self, other):
        if isinstance(other, Matrix):
            if self.rows == other.rows and self.cols == other.cols:
                self.data = [row[:] for row in other.data]
            else:
                raise ValueError("Matrix sizes do not match")
        elif isinstance(other, list):
            if len(other) == self.rows and all(len(row) == self.cols for row in other):
                self.data = [row[:] for row in other]
            else:
                raise ValueError("List dimensions do not match matrix size")
        else:
            raise TypeError("Assignment must be from a Matrix or a list of lists")
    
    def shape(self):
        return (self.rows, self.cols)
    
    def transpose(self):
        return Matrix([[self.data[j][i] for j in range(self.rows)] for i in range(self.cols)])
    
    def row(self, n):
        return Matrix([self.data[n]])
    
    def column(self, n):
        return Matrix([[self.data[i][n]] for i in range(self.rows)])
    
    def to_list(self):
        return [row[:] for row in self.data]
    
    def block(self, n_0, n_1, m_0, m_1):
        return Matrix([row[m_0:m_1] for row in self.data[n_0:n_1]])

# Testing the implementation
if __name__ == "__main__":
    print("Initializing a 3x3 zero matrix:")
    M1 = Matrix(3, 3)
    print(M1, "\n")
    
    print("Initializing a matrix with a list of lists:")
    M2 = Matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
    print(M2, "\n")
    
    print("Testing indexing:")
    print("M2[1,2] =", M2[1, 2])
    print("M2[2] =", M2[2], "\n")
    
    print("Testing assignment:")
    M1.assign(M2)
    print("M1 after assignment:")
    print(M1, "\n")
    
    print("Testing element modification:")
    M1[0, 0] = 10
    print(M1, "\n")
    
    print("Testing invalid assignment:")
    try:
        M1.assign([[1, 2], [3, 4]])
    except ValueError as e:
        print("Error:", e)
    
    print("Testing shape:")
    print("Shape of M2:", M2.shape(), "\n")
    
    print("Testing transpose:")
    print(M2.transpose(), "\n")
    
    print("Testing row extraction:")
    print(M2.row(1), "\n")
    
    print("Testing column extraction:")
    print(M2.column(1), "\n")
    
    print("Testing to_list method:")
    print(M2.to_list(), "\n")
    
    print("Testing block extraction:")
    print(M2.block(0, 2, 1, 3), "\n")
```

4. Add the following member functions to your class. Make sure to appropriately test the dimensions of the matrices to make sure the operations are correct.
    * `M.scalarmul(c)`: a matrix that is scalar product $cM$, where every element of $M$ is multiplied by $c$.
    * `M.add(N)`: adds two matrices $M$ and $N$. Don’t forget to test that the sizes of the matrices are compatible for this and all other operations.
    * `M.sub(N)`: subtracts two matrices $M$ and $N$.
    * `M.mat_mult(N)`: returns a matrix that is the matrix product of two matrices $M$ and $N$.
    * `M.element_mult(N)`: returns a matrix that is the element-wise product of two matrices $M$ and $N$.
    * `M.equals(N)`: returns true/false if $M==N$.

```python
class Matrix:
    def __init__(self, n, m=None):
        if isinstance(n, list):  # Initialize with a list of lists
            self.data = n
            self.rows = len(n)
            self.cols = len(n[0]) if n else 0
            if any(len(row) != self.cols for row in n):
                raise ValueError("All rows must have the same number of columns")
        else:  # Initialize with dimensions
            self.rows = n
            self.cols = m
            self.data = [[0] * m for _ in range(n)]

    def __getitem__(self, index):
        if isinstance(index, tuple):
            i, j = index
            if isinstance(i, slice) or isinstance(j, slice):
                return Matrix([row[j] for row in self.data[i]])
            return self.data[i][j]
        return self.data[index]

    def __setitem__(self, index, value):
        if isinstance(index, tuple):
            i, j = index
            self.data[i][j] = value
        else:
            self.data[index] = value

    def __repr__(self):
        return '\n'.join(str(row) for row in self.data)
    
    def assign(self, other):
        if isinstance(other, Matrix):
            if self.rows == other.rows and self.cols == other.cols:
                self.data = [row[:] for row in other.data]
            else:
                raise ValueError("Matrix sizes do not match")
        elif isinstance(other, list):
            if len(other) == self.rows and all(len(row) == self.cols for row in other):
                self.data = [row[:] for row in other]
            else:
                raise ValueError("List dimensions do not match matrix size")
        else:
            raise TypeError("Assignment must be from a Matrix or a list of lists")
    
    def shape(self):
        return (self.rows, self.cols)
    
    def transpose(self):
        return Matrix([[self.data[j][i] for j in range(self.rows)] for i in range(self.cols)])
    
    def row(self, n):
        return Matrix([self.data[n]])
    
    def column(self, n):
        return Matrix([[self.data[i][n]] for i in range(self.rows)])
    
    def to_list(self):
        return [row[:] for row in self.data]
    
    def block(self, n_0, n_1, m_0, m_1):
        return Matrix([row[m_0:m_1] for row in self.data[n_0:n_1]])
    
    def scalarmul(self, c):
        return Matrix([[c * self.data[i][j] for j in range(self.cols)] for i in range(self.rows)])
    
    def add(self, N):
        if self.shape() != N.shape():
            raise ValueError("Matrix sizes do not match for addition")
        return Matrix([[self.data[i][j] + N.data[i][j] for j in range(self.cols)] for i in range(self.rows)])
    
    def sub(self, N):
        if self.shape() != N.shape():
            raise ValueError("Matrix sizes do not match for subtraction")
        return Matrix([[self.data[i][j] - N.data[i][j] for j in range(self.cols)] for i in range(self.rows)])
    
    def mat_mult(self, N):
        if self.cols != N.rows:
            raise ValueError("Matrix sizes do not match for multiplication")
        return Matrix([[sum(self.data[i][k] * N.data[k][j] for k in range(self.cols)) for j in range(N.cols)] for i in range(self.rows)])
    
    def element_mult(self, N):
        if self.shape() != N.shape():
            raise ValueError("Matrix sizes do not match for element-wise multiplication")
        return Matrix([[self.data[i][j] * N.data[i][j] for j in range(self.cols)] for i in range(self.rows)])
    
    def equals(self, N):
        return self.shape() == N.shape() and all(self.data[i][j] == N.data[i][j] for i in range(self.rows) for j in range(self.cols))

def constant(n, m, c):
    return Matrix([[float(c)] * m for _ in range(n)])

def zeros(n, m):
    return constant(n, m, 0.0)

def ones(n, m):
    return constant(n, m, 1.0)

def eye(n):
    return Matrix([[1.0 if i == j else 0.0 for j in range(n)] for i in range(n)])
```

```python
# Creating matrices
A = Matrix([[1, 2], [3, 4]])
B = Matrix([[5, 6], [7, 8]])

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

# Scalar multiplication
C = A.scalarmul(2)
print("\nA * 2:")
print(C)

# Matrix addition
D = A.add(B)
print("\nA + B:")
print(D)

# Matrix subtraction
E = A.sub(B)
print("\nA - B:")
print(E)

# Matrix multiplication
F = A.mat_mult(B)
print("\nA * B:")
print(F)

# Element-wise multiplication
G = A.element_mult(B)
print("\nA .* B (element-wise multiplication):")
print(G)

# Transpose of A
H = A.transpose()
print("\nTranspose of A:")
print(H)

# Identity matrix
I = eye(3)
print("\nIdentity Matrix (3x3):")
print(I)

# Checking equality
print("\nA == B:", A.equals(B))
print("A == A:", A.equals(A))

# Accessing row and column
print("\nRow 1 of A:")
print(A.row(1))

print("\nColumn 0 of A:")
print(A.column(0))
```

```python
class Matrix:
    def __init__(self, n, m=None):
        if isinstance(n, list):  # Initialize with a list of lists
            self.data = n
            self.rows = len(n)
            self.cols = len(n[0]) if n else 0
            if any(len(row) != self.cols for row in n):
                raise ValueError("All rows must have the same number of columns")
        else:  # Initialize with dimensions
            self.rows = n
            self.cols = m
            self.data = [[0] * m for _ in range(n)]

    def __getitem__(self, index):
        if isinstance(index, tuple):
            i, j = index
            if isinstance(i, slice) or isinstance(j, slice):
                return Matrix([row[j] for row in self.data[i]])
            return self.data[i][j]
        return self.data[index]

    def __setitem__(self, index, value):
        if isinstance(index, tuple):
            i, j = index
            self.data[i][j] = value
        else:
            self.data[index] = value

    def __repr__(self):
        return '\n'.join(str(row) for row in self.data)
    
    def __mul__(self, other):
        if isinstance(other, (int, float)):
            return self.scalarmul(other)
        elif isinstance(other, Matrix):
            return self.mat_mult(other)
        else:
            raise TypeError("Unsupported operand type for *: 'Matrix' and '{}'".format(type(other)))
    
    def __rmul__(self, other):
        return self.__mul__(other)
    
    def __add__(self, other):
        return self.add(other)
    
    def __sub__(self, other):
        return self.sub(other)
    
    def __eq__(self, other):
        return self.equals(other)
    
    def assign(self, other):
        if isinstance(other, Matrix):
            if self.rows == other.rows and self.cols == other.cols:
                self.data = [row[:] for row in other.data]
            else:
                raise ValueError("Matrix sizes do not match")
        elif isinstance(other, list):
            if len(other) == self.rows and all(len(row) == self.cols for row in other):
                self.data = [row[:] for row in other]
            else:
                raise ValueError("List dimensions do not match matrix size")
        else:
            raise TypeError("Assignment must be from a Matrix or a list of lists")
    
    def shape(self):
        return (self.rows, self.cols)
    
    def transpose(self):
        return Matrix([[self.data[j][i] for j in range(self.rows)] for i in range(self.cols)])
    
    def row(self, n):
        return Matrix([self.data[n]])
    
    def column(self, n):
        return Matrix([[self.data[i][n]] for i in range(self.rows)])
    
    def to_list(self):
        return [row[:] for row in self.data]
    
    def block(self, n_0, n_1, m_0, m_1):
        return Matrix([row[m_0:m_1] for row in self.data[n_0:n_1]])
    
    def scalarmul(self, c):
        return Matrix([[c * self.data[i][j] for j in range(self.cols)] for i in range(self.rows)])
    
    def add(self, N):
        if self.shape() != N.shape():
            raise ValueError("Matrix sizes do not match for addition")
        return Matrix([[self.data[i][j] + N.data[i][j] for j in range(self.cols)] for i in range(self.rows)])
    
    def sub(self, N):
        if self.shape() != N.shape():
            raise ValueError("Matrix sizes do not match for subtraction")
        return Matrix([[self.data[i][j] - N.data[i][j] for j in range(self.cols)] for i in range(self.rows)])
    
    def mat_mult(self, N):
        if self.cols != N.rows:
            raise ValueError("Matrix sizes do not match for multiplication")
        return Matrix([[sum(self.data[i][k] * N.data[k][j] for k in range(self.cols)) for j in range(N.cols)] for i in range(self.rows)])
    
    def element_mult(self, N):
        if self.shape() != N.shape():
            raise ValueError("Matrix sizes do not match for element-wise multiplication")
        return Matrix([[self.data[i][j] * N.data[i][j] for j in range(self.cols)] for i in range(self.rows)])
    
    def equals(self, N):
        return self.shape() == N.shape() and all(self.data[i][j] == N.data[i][j] for i in range(self.rows) for j in range(self.cols))

def constant(n, m, c):
    return Matrix([[float(c)] * m for _ in range(n)])

def zeros(n, m):
    return constant(n, m, 0.0)

def ones(n, m):
    return constant(n, m, 1.0)

def eye(n):
    return Matrix([[1.0 if i == j else 0.0 for j in range(n)] for i in range(n)])
```

6. Demonstrate the basic properties of matrices with your matrix class by creating two 2 by 2 example matrices using your Matrix class and illustrating the following:

$$
(AB)C=A(BC)
$$
$$
A(B+C)=AB+AC
$$
$$
AB\neq BA
$$
$$
AI=A
$$

```python
# Creating example matrices
A = Matrix([[1, 2], [3, 4]])
B = Matrix([[2, 0], [1, 3]])
C = Matrix([[0, 1], [4, 2]])
I = eye(2)  # 2x2 Identity matrix

# Associativity of matrix multiplication: (AB)C = A(BC)
AB = A * B
BC = B * C
left = AB * C
right = A * BC

print("(AB)C:")
print(left)
print("\nA(BC):")
print(right)
print("\n(AB)C == A(BC):", left == right)

# Distributive property: A(B + C) = AB + AC
B_plus_C = B + C
left = A * B_plus_C
right = (A * B) + (A * C)

print("\nA(B + C):")
print(left)
print("\nAB + AC:")
print(right)
print("\nA(B + C) == AB + AC:", left == right)

# Non-commutativity: AB ≠ BA
BA = B * A

print("\nAB:")
print(AB)
print("\nBA:")
print(BA)
print("\nAB == BA:", AB == BA)

# Identity property: AI = A
AI = A * I

print("\nAI:")
print(AI)
print("\nA:")
print(A)
print("\nAI == A:", AI == A)
```
